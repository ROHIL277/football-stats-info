# **Football Stats & Info**

This notebook depicts some basic EDA done on a sample dataset from Kaggle and it there is exploration of visualisation libraries like seaborn , matplot lib in this notebook

**Note: This is just a sample data and this doesn't pertain to a stand alone EPL Season per se**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv(r'C:\Kaggle\football_dataset.csv')
df
```

```python
df.info
```

```python
df.dtypes
```

```python
df.columns
```

```python
selected_df = df[['Name', 'Passes', 'Assists','Appearances']]
selected_df['Passes/Assist Ratio'] = (selected_df['Passes'])/(selected_df['Assists']*selected_df['Appearances'])
filtered_df = selected_df[selected_df['Passes/Assist Ratio'] != float('inf')]
filtered_df
```

```python
top_10_players_pass_assist = filtered_df.sort_values(by='Passes/Assist Ratio', ascending=False).head(10)
top_10_players_pass_assist
```

```python
top_10_players = df.sort_values(by='Goals', ascending=False).head(10)
print(top_10_players)
```

```python
df_players =  df[['Name', 'Goals', 'Assists','Passes','Tackles']]
df_players
```

```python
top_10_players
```

```python
df_goalkeeper = df[df['Position'] == "Goalkeeper"]
df_goalkeeper
```

```python
df_defender = df[df['Position'] == "Defender"]
df_defender
```

```python
df_defender['defender_quality'] = (df_defender['Goals'] + df_defender['Tackles'] + df_defender['Interceptions']+df_defender['Clearances']+df_defender['Duels won']+df_defender['Aerial battles won']-df_defender['Duels lost']-df_defender['Aerial battles lost']-df_defender['Own goals'])/df_defender['Appearances']
df_defender
```

```python
top_10_defenders = df_defender.sort_values(by='defender_quality', ascending=False).head(10)
top_10_defenders
```

```python
df_goalkeeper['saves_per_match'] = df_goalkeeper['Saves'] / df_goalkeeper['Appearances']
df_goalkeeper
```

```python
top_10_keepers = df_goalkeeper.sort_values(by='saves_per_match', ascending=False).head(10)
top_10_keepers
```

## **Heatmap -Correlation Matrix of Player Stats**

```python
numeric_cols = ['Goals', 'Assists', 'Passes', 'Tackles']
corr_matrix = df_players[numeric_cols].corr()
corr_matrix
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Player Stats')
plt.show()
```

## **Key Summary Stats**

```python
summary_stats = df[numeric_cols].agg(['mean', 'median', 'min', 'max', 'std'])
summary_stats
```

## **Top 10 Goal Keepers (Saves per match)**

```python
plt.figure(figsize=(8, 5))
palette = sns.color_palette("crest",10)[::-1]
sns.barplot(x='saves_per_match',y='Name',data = top_10_keepers,palette=palette)
plt.title('Top 10 Goal Keepers')
plt.show()
```

## **Top 10 Goal Scorers**

```python
plt.figure(figsize=(6, 4))
palette = sns.color_palette("crest",10)[::-1]
sns.barplot(x='Goals',y='Name',data = top_10_players,palette=palette)
plt.title('Top 10 Goal Scorers')
plt.show()
```

## **Top 10 Defenders(cummalitive factors)**

```python
plt.figure(figsize=(6, 4))
palette = sns.color_palette("crest",10)[::-1]
sns.barplot(x='defender_quality',y='Name',data = top_10_defenders,palette=palette)
plt.title('Top 10 Defenders')
plt.show()
```

# **Top 10 Players - Pass/Assist**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 10))

# Scatter plot
sns.scatterplot(data=top_10_players_pass_assist, x='Appearances', y='Passes/Assist Ratio')

# Add player names above each point
for i in range(len(top_10_players_pass_assist)):
    plt.text(
        top_10_players_pass_assist['Appearances'].iloc[i],
        top_10_players_pass_assist['Passes/Assist Ratio'].iloc[i] + 0.2,
        top_10_players_pass_assist['Name'].iloc[i],
        fontsize=9,
        ha='center',
        va='bottom'
    )

plt.title('Passes vs Passes/Assist Ratio')
plt.xlabel('Appearances')
plt.ylabel('Passes/Assist Ratio per match')

# REMOVE this line to get rid of grid
# plt.grid(True)

plt.show()


```
